```python
# combine city and state
def combine(df):
    state=df["State"]
    city=df["City"]
    return f"{city}, {state}"
df["City, State"]=df.apply(combine, axis=1)
```

```python
# https://stackoverflow.com/questions/39512002/convert-whole-dataframe-from-lower-case-to-upper-case-with-pandas
# merge df with Ref tables
City=City.apply(lambda x: x.astype(str).str.upper())
Manager=Manager.apply(lambda x: x.astype(str).str.upper())
Rep=Rep.apply(lambda x: x.astype(str).str.upper())
```

```python
# this returns duplicate columns
df.merge(City, left_on='City, State', right_on='City')
```

```python
# use index instead
# https://stackoverflow.com/questions/36538780/merging-dataframes-on-index-with-pandas
df.set_index("City, State").merge(City.set_index("City"), how='outer', left_index=True, right_index=True)
```

```python
# define funtion to merge index
def MergeIndex(df1,df2,index1,index2):
    df1=df1.set_index(index1)
    df2=df2.set_index(index2)
    mergedDF=df1.merge(df2, how='outer', left_index=True, right_index=True)
    return mergedDF.reset_index().rename(columns={"index": index1})

df=MergeIndex(df,City,"City, State","City")
df=MergeIndex(df,Manager,"Region","Region")
df=MergeIndex(df,Rep,"City, State","City")
```

```python
# convert string to floats
# series are of object type instead of numbers
df["Investment"]
```

```python
def convertToFloat(currency):
    for ch in ['$',',']:
        if ch in currency:
            currency = currency.replace(ch,"")
    return float(currency)
```

```python
df["Investment"]=df["Investment"].apply(convertToFloat)
df["Revenue"]=df["Revenue"].apply(convertToFloat)
```

```python
df.head()
```

```python
# convert date from string to datetime object
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df["Date"]
```

```python
df["Date"].iloc[0]
```

```python
# get net income
df["Net Income"]=df["Revenue"]-df["Investment"]
```

```python
df
```

```python
# determine if there was a profit or loss
df["Profit/Loss"]=df["Net Income"].apply(lambda x:"Profit" if x>0 else "Loss")
```

```python
# determine retrn on interest
def ROI(df):
    a=df["Investment"]
    b=df["Net Income"]
    return round((b/a),3)

df["ROI"]=df.apply(ROI, axis=1)
```

```python
df
```

```python
# get month year and quarter
df["year"]=df["Date"].apply(lambda x:x.year)
df["month"]=df["Date"].apply(lambda x:x.month)
df["quarter"]=df["Date"].apply(lambda x:f"Q {x.quarter}")
```

```python
# create heat map:
#     year and month
#     year and region
#     year and city
# groupby objects:
#     year quarter month profit/loss

df.groupby(["year","quarter","month","Region"])["ROI"].mean().to_frame()
```

```python
def getHeatMap(df,indexList,columnList, axis):
    HeatMap=pd.pivot_table(df, values = 'ROI', index=indexList, columns=columnList)
    HeatMap=HeatMap.round(3).style.background_gradient(cmap=sns.light_palette("green", as_cmap=True),axis=axis)
    return HeatMap
```

```python
# get year by region
yearRegion=getHeatMap(df,
           indexList=["year"],
           columnList=['Region'],axis=1)
```

```python
# get year by region
yearProperty=getHeatMap(df,
           indexList=["year"],
           columnList=['Property Type'],axis=1)
```

```python
# get year by month
yearMonth=getHeatMap(df,indexList=["year"], columnList=['month'],axis=1)
```

```python
# get city by year
cityYear=getHeatMap(df,indexList=["City, State"], columnList=["year"],axis=0)
```

```python
import calendar
def getSummary(subtype):
    d1={}
    for year in df["year"].unique():
        for region in df["Region"].unique():
            DF=df[(df["year"]==year) & (df["Region"]==region)]
            if subtype=="A":
                DF=DF.groupby(["quarter","Profit/Loss"])["Profit/Loss"].count().unstack()
                DF=DF[["Profit","Loss"]]
            elif subtype=="B":
                DF=pd.pivot_table(DF, values = 'ROI', index=["quarter"], columns = 'Region')
            elif subtype=="C":
                DF=pd.pivot_table(DF, values = 'ROI', index=["month"], columns = 'Region')
                DF.index=DF.index.map(lambda x:calendar.month_abbr[x])
            DF=DF.round(3)
            d1[f"{year}-{region}"]=DF
    return d1
```

```python
def InsertTable(_DICTIONARY,_WRITER,title):
    row=0
    for i,x in enumerate(_DICTIONARY):
        tempDF=_DICTIONARY[x]
        sLen=len(tempDF[tempDF.columns[0]])
        tempDF.to_excel(_WRITER,
                        sheet_name = title
                        ,startrow=row, startcol=0)
        row=(row+sLen+2)
```

```python
def InsertChart(_DICTIONARY,
                _WORKSHEET,
                WORKBOOK,
                _SheetName):
    chart_List=[]
    _Row=1
    _Columns=0
    for i,x in enumerate(_DICTIONARY):
        title=f"{x}"
        tempDF=_DICTIONARY[x]
        qaz=tempDF.iloc[0]
        sLen=len(qaz)+1
        rLen=len(list(_DICTIONARY[x].index))
        chart = WORKBOOK.add_chart({'type': 'column'})
        for col_num in range(1, sLen):
            chart.add_series({
                'name':       [_SheetName, _Columns, col_num],
                'categories': [_SheetName, _Row, 0, (rLen+_Row-1), 0],
                'values':     [_SheetName, _Row, col_num, (rLen+_Row-1), col_num],
                'overlap':    -5,
                'data_labels': {'value': True,
                                'font': {'name': 'Calibri', "bold":True, "size":10}}
            })
            chart.set_x_axis({'name':"", 'major_gridlines': {'visible': False}})
            chart.set_y_axis({'name':"", 'major_gridlines': {'visible': False}})
        chart.set_title ({'name':title})
        chart.set_legend({'position': 'none'})
        chart.set_size({'width': 700, 'height': 477.252})
        chart_List.append(chart)
        _Row =_Row+rLen+2
        _Columns = _Columns+rLen+2
    # ------------------------------------------------------------------------------------
    # Insert the chart into the worksheet.
    num1=1
    num2=5
    _count_=0
    cellDict={1: 'A', 2: 'L', 3: 'W', 4: 'AH', 5: 'AS', 6: 'BD', 7: 'BO'}
    for i,x in enumerate(_DICTIONARY):
        _count_+=1
        _WORKSHEET.insert_chart(cellDict[_count_]+str(num1), chart_List[i],{"x_scale":1,"y_scale":1})
        if _count_ == 7:
            _count_= 0
            num1+=24
```

```python
def SaveChartsToFile(dictList,savePath):
    CurrentDate=datetime.datetime.now().date()
    with pd.ExcelWriter(f"{savePath}\ROI [{CurrentDate}].xlsx") as writer:
#         write heat maps
        tabList=["Year_PropertyType","Year_Region",
                 "Year_Month","City_Year"]
        dfList=[yearProperty,yearRegion,yearMonth,cityYear]
        for a,b in zip(dfList,tabList):
            a.to_excel(writer,sheet_name=b)
            
#         write charts
        workbook  = writer.book
        Dashboard_1 = workbook.add_worksheet("Profit_Loss by year")
        Dashboard_2 = workbook.add_worksheet("Avg ROI by quarter")
        Dashboard_3 = workbook.add_worksheet("Avg ROI by month")
        DashboardList=[Dashboard_1,Dashboard_2,Dashboard_3]
        for x in DashboardList:
            x.set_zoom(36)
        RefList=["Ref 1","Ref 2","Ref 3"]
        for a,b,c in zip(dictList,DashboardList,RefList):
            InsertTable(a,writer,c)
            InsertChart(a,b,workbook,c)  
```

```python
savePath=r"C:\Users\immanuel\Desktop\FEED MIGRATION\Data Migration\Sales Data\Summary"
dictList=[getSummary("A"),getSummary("B"),getSummary("C")]
SaveChartsToFile(dictList,savePath)
```
